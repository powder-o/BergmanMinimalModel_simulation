```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def odes(y,t,args=params):

    p1,p2,p3,p4,p5,p6=params
    
    G = y[0]
    X = y[1]
    I = y[2]
    
    dGdt = -(p1 + X)*G + (p1*Gb) 
    dXdt = -(p2*X) + p3*(I-Ib)
    dIdt = p6*(G-p5)*t - p4*(I-Ib)
    if(G<p5):
        dIdt = -p4*(I-Ib)
    
    #return [dGdt,dIdt]                   
    return [dGdt,dXdt,dIdt]
```

```python
params=(0.03082,
0.014,
1.062e-5,
0.3,
100,
0.003349)

p1,p2,p3,p4,p5,p6=params
```

```python
Gb=90
Ib=7.3
Xb=0
G0=100          
y0 = [Gb+G0, Xb, Ib]
```

```python
t=np.linspace(0,360,361)
```

```python
y=odeint(odes,y0,t)
```

```python
G=y[:,0]
X=y[:,1]
I=y[:,2]
```

```python
plt.axhline(y = Gb, color = 'g', linestyle = '--',label="Gb")
#plt.axhline(y = Ib, color = 'k', linestyle = '--',label="Ib")

plt.plot(t,G,'g',label="G")
#plt.plot(t,I,'r',label="I")

plt.legend()
plt.show()
```

```python
plt.axhline(y = Xb, color = 'k', linestyle = '--',label="Xb")
plt.plot(t,X,'b',label="X")
plt.legend()
plt.show()
```

```python
plt.plot(t,I,'r',label="I(t)")
plt.axhline(y = Ib, color = 'k', linestyle = '--',label="Ib")
plt.legend()
plt.show()
```

```python
I_max = np.amax(I)
print("I_max: ",I_max)

phi_1 = (I_max - Ib)/(p4*(G0-Gb))
phi_2 = p6*(10**4)

SI=p3/p2
AIRg=phi_2

print(f'PHI_1 = {phi_1}\nPHI_2 = {phi_2}\nSI = {SI}\nAIRg = {AIRg}')
print(AIRg*SI)
```
